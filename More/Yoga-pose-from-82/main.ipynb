## Count images

```python
import os
import openpyxl
import re

def count_images_and_create_excel(source_folder, excel_file_name):
    # Tạo một Workbook mới
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.append(['Class name', 'Số lượng ảnh'])
    # Lấy danh sách thư mục con và sắp xếp chúng theo thứ tự
    subfolders = os.listdir(source_folder)
    subfolders.sort(key=lambda x: int(re.sub('\D', '', x)))
    for folder_class in subfolders:
        folder_img = os.path.join(source_folder, folder_class)
        file_count = 0
        for root, dirs, files in os.walk(folder_img):
            for file in files:
                file_count += 1
        print(f'{folder_class} : {file_count}')
        sheet.append([folder_class, file_count])
    workbook.save(excel_file_name)
    print("-"*100)
    print(f"File '{excel_file_name}' saved successfully!")

def count_images_and_create_excel_sub(source_folder, excel_file_name):
    # Tạo một Workbook mới
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.append(['Class name', 'Số lượng ảnh'])
    for folder_class in os.listdir(source_folder):
        folder_img = os.path.join(source_folder, folder_class)
        file_count = 0
        for root, dirs, files in os.walk(folder_img):
            for file in files:
                file_count += 1
        print(f'{folder_class} : {file_count}')
        sheet.append([folder_class, file_count])
    workbook.save(excel_file_name)
    print("-"*100)
    print(f"File '{excel_file_name}' saved successfully!")
```

# Make csv 6 20

```python
boundaries_yg6 = [0, 20, 38, 46, 53, 72, len(list_class_82)]
boundaries_yg20 = [0, 3, 8, 15, 20, 25, 29, 31, 35, 38, 43, 46, 51, 53, 61, 66, 68, 72, 79, 80, len(list_class_82)]
def make_class_mapping(list_class_82, list_folder, boundaries):
    class_name_mapping = {}
    for i in range(1, len(boundaries)):
        start = boundaries[i-1]
        end = boundaries[i]
        folder_name = list_folder[i-1]

        for class_name in list_class_82[start:end]:
            class_name_mapping[class_name] = folder_name
    return class_name_mapping
            
class_name_mapping_yg_6 = make_class_mapping(list_class_82, list_folder_6, boundaries_yg6)
class_name_mapping_yg_20 = make_class_mapping(list_class_82, list_folder_20, boundaries_yg20)
```

```python
def make_yg_csv(csv_path, class_name_mapping, list_folder):
    df = pd.read_csv(csv_path)
    df['class_name'] = df['class_name'].replace(class_name_mapping)
    for i in range(len(list_folder)):
        df.loc[df['class_name'] == list_folder[i], 'class_no'] = i
    df = df.sort_values(by='class_no')
    df.to_csv(f"{csv_path.split('.')[0]}_yg{len(list_folder)}.csv", index = False)
    
    return df
```

# Make pose per class

```python
import pandas as pd
import os

def make_pose_csv(csv_path, path_out):
    if not os.path.exists(path_out):
        os.makedirs(path_out)
    df = pd.read_csv(csv_path)
    class_names = df['class_name'].unique()
    for class_name in class_names:
        class_dataframes = df[df['class_name'] == class_name]
        class_dataframes.to_csv(f"{path_out}/{class_name}.csv", index = False)
        print(f'Tạo thành công {path_out}/{class_name}.csv')
        
path_out_train = 'data_per_pose_train'
path_out_test = 'data_per_pose_test'
path_data = 'data'        
csv_path_train = f"{path_data}/train_data.csv"
csv_path_test = f"{path_data}/test_data.csv"

make_pose_csv(csv_path_train, path_out_train)
make_pose_csv(csv_path_test, path_out_test)
```

```python
import pandas as pd
import os

def concat_train_test_csv(path_train, path_test, path_out):
    if not os.path.exists(path_out):
        os.makedirs(path_out)
    df_train = pd.read_csv(path_train)
    df_test = pd.read_csv(path_test)
    df_test = df_test.iloc[0:] #Loại bỏ heading
    class_names = df_train['class_name'].unique()
    result = pd.concat([df_train, df_test])
    
    result.to_csv(f"{path_out}/{class_names[0]}.csv", index = False)
    print(f'Tạo thành công {path_out}/{class_names[0]}.csv')
        
path_out = 'data_per_pose'
path_data_train = 'data_per_pose_train'
path_data_test = 'data_per_pose_test'
list_train = []
list_test = []

for path_pose_train in os.listdir(path_data_train):
    pose_train = os.path.join(path_data_train, path_pose_train)
    list_train.append(pose_train)
for path_pose_test in os.listdir(path_data_test):
    pose_test = os.path.join(path_data_test, path_pose_test)
    list_test.append(pose_test)
    
for i in range(len(list_train)):
    path_train = list_train[i]
    path_test = list_test[i]
    concat_train_test_csv(path_train, path_test, path_out)
```
